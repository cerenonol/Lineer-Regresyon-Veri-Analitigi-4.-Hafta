```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

data = {
    'Ülke':['Almanya','İspanya','Hollanda','Danimarka','Belçika',
               'İtalya','İsviçre','İrlanda','Fransa','Portekiz',
               'Türkiye','İran','Kırgısiztan','Tanzanya','Polonya'],
    'Haftalık_Ortalama_Saatler':[34,36,34,34,34,
                                 36,36,36,36,36,
                                 44,46,40,44,39],
    'Refah_Düzeyi':[6892,6436,6892,7586,6859,
                    6405,7240,6911,6661,5968,
                    4614,4876,5825,3694,6260],
    'Kişi_Başına_Gelir':[52722,33071,62719,68300,53659,
                         38326,100413,104272,46001,27880,
                         12849,4663,1843,1254,21996]
}

df = pd.DataFrame(data)

X=df[['Haftalık_Ortalama_Saatler','Refah_Düzeyi']]
y=df['Kişi_Başına_Gelir']

model=LinearRegression()
model.fit(X,y)

tahminler=model.predict(X)

mse=mean_squared_error(y,tahminler)
r_squared=r2_score(y,tahminler)
print(f'Ortalama Kare Hatası : {mse}')
print(f'R-Kare Değeri : {r_squared}')

plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
sns.scatterplot(x=df['Haftalık_Ortalama_Saatler'],y=y,label='Kişi_Başına_Gelir',color='blue')
sns.regplot(x=df['Haftalık_Ortalama_Saatler'],y=y,scatter=False,color='red',line_kws={"label":"Doğrusal Uygunluk"})
plt.title('Kişi Başına Gelir - Haftalık Ortalama Saatler')
plt.xlabel('Haftalık_Ortalama_Saatler')
plt.ylabel('Kişi_Başına_Gelir')
plt.legend()

plt.subplot(1,2,2)
sns.scatterplot(x=df['Refah_Düzeyi'],y=y,label='Kişi_Başına_Gelir',color='green')
sns.regplot(x=df['Refah_Düzeyi'],y=y,scatter=False,color='orange',line_kws={"label":"Doğrusal Uygunluk"})
plt.title('Kişi Başına Gelir - Refah_Düzeyi')
plt.xlabel('Refah_Düzeyi')
plt.ylabel('Kişi_Başına_Gelir')
plt.legend()

plt.tight_layout()
plt.show()







```
